### Plota a resposta em frequencia de um media movel


```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.signal import freqz
```

```python
## Variaveis
passo = np.pi/1000;
#w = 0:passo:pi;
w = np.arange(0, np.pi, passo)
L = 8;
Fs = 8000;
```

```python
# % Numerador e Denominador
num = np.sin(w*L/2);
den = np.sin(w/2);

#% Equação Base
temp = num/den;

```

```python
# Plotando a frequencia em Rad

X = (1/L) *( np.abs(temp));

#subplot(3,1,1);
plt.figure(figsize = (12, 12))
plt.subplot(3,1,1)
plt.plot(w,X);
plt.grid()
plt.xlabel('Frequência');
plt.title('Frequência Rad');
#grid on;

```

```python
# Plotando a frequencia em Hz
F_Hz = (w/np.pi)*(Fs/2);

plt.figure(figsize = (12, 12))
plt.subplot(3,1,2);
plt.plot(F_Hz,X);
plt.xlabel('Frequência');
plt.title('Frequência em Hz');
plt.grid();
```

```python
# Plotando a Magnitude em Db
X_Db = 20* np.log10 (X);

plt.figure(figsize = (12, 12))
plt.subplot(3,1,3);
plt.plot(F_Hz,X_Db);
plt.ylabel('Atenuação DB');
plt.xlabel('Frequência');
plt.title('Frequência em Hz');
plt.grid();

plt.ylim(-70,0) # Redefinindo o eixo y

#plt.axis([0 4000 -70 0]);
```

```python
# Usando o Freqz
num_ = np.zeros((1, L), dtype='float64')
num_[0, :] = 1 / L

den_ = float(1)
num_
```

```python
[w, h] = freqz(num_.T, den_, worN=Fs, fs=Fs)

plt.figure(figsize = (12, 12))
plt.plot(w, 20 * np.log10(abs(h)), 'b')

plt.title('Magnitude da resposta em frequencia')
plt.grid()

#show()
```
